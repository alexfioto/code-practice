# Limits

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(-25, 25):
    if i != 1:
        plt.scatter(i, 1, marker='_', color='blue')
    else:
        plt.scatter(i, 1, marker='*', color='red')
```

```python
lst = np.array(range(-100, 100))
```

```python
parabola = lst ** 2
```

```python
plt.scatter(lst, parabola, marker='.');
```

```python
new_parabola = np.where(lst == 50, 1, parabola )
```

```python
plt.scatter(lst, new_parabola, marker='.')
```
